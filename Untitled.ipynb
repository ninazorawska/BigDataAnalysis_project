```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder \
    .appName("MySparkApp") \
    .config("spark.driver.memory", "2g") \
    .getOrCreate()

spark
```

```python
data = [("Alice", 23), ("Bob", 34), ("Cathy", 29)]
columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)
df.show()

```

```python
df = spark.read.csv("wine+quality/winequality-red.csv", sep = ';', header=True, inferSchema=True)
df.show()
```
